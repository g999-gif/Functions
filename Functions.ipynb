#Functions

1. Difference between a function and a method in Python:
- A function is a standalone piece of reusable code, while a method is associated with an object and can access its data.

2. Explain function arguments and parameters in Python:

- Parameters are placeholders defined in a function (e.g., def func(x):).
- Arguments are the actual values passed to a function when calling it (e.g., func(5)).

3. Ways to define and call a function in Python:
- Define using def or lambda. Call it by using its name followed by parentheses.
-def greet(): return "Hello"
print(greet())

4. Purpose of the return statement in a Python function:
- To send a result from a function back to its caller.
-def add(a, b): return a + b
print(add(3, 4))  # Output: 7

5. Iterators vs Iterables in Python:

- Iterable: An object you can loop over (e.g., list, string).
- Iterator: An object that produces values one at a time using __next__().
- nums = iter([1, 2, 3])  # nums is an iterator
print(next(nums))  # Output: 1

6. Concept of generators in Python:
- Generators produce items lazily using yield and save memory.
- def gen(): yield 1; yield 2
print(list(gen()))  # Output: [1, 2]

7. Advantages of using generators over functions:

- Lower memory usage.
- Efficient for large datasets.
- Produces values on-demand.

8. Lambda function in Python:
- Anonymous one-liner functions used for short tasks.
- square = lambda x: x**2
print(square(4))  # Output: 16

9. Purpose and usage of map():
- Applies a function to all items in an iterable.
- nums = [1, 2, 3]
result = map(lambda x: x**2, nums)
print(list(result))  # Output: [1, 4, 9]

10. Difference between map(), reduce(), and filter():

- map(): Transforms each item.
- reduce(): Reduces iterable to a single value (requires functools).
- filter(): Filters items based on a condition.

```python
# 1. Sum of all even numbers in a list:
def sum_even(nums):
    return sum(x for x in nums if x % 2 == 0)

print(sum_even([1, 2, 3, 4]))  # Output: 6
```

```python
# 2. Reverse a string:
def reverse_string(s):
    return s[::-1]

print(reverse_string("hello"))  # Output: "olleh"
```

```python
# 3. Squares of numbers in a list:
def square_list(nums):
    return [x**2 for x in nums]

print(square_list([1, 2, 3]))  # Output: [1, 4, 9]
```

```python
# 4. Check if a number is prime (1200):
def is_prime(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0: return False
    return True

primes = [x for x in range(1, 201) if is_prime(x)]
print(primes)
```

```python
# 5. Iterator class for Fibonacci sequence:
class Fibonacci:
    def __init__(self, n): self.n, self.a, self.b = n, 0, 1
    def __iter__(self): return self
    def __next__(self):
        if self.n <= 0: raise StopIteration
        self.n -= 1
        fib, self.a, self.b = self.a, self.b, self.a + self.b
        return fib

fib = Fibonacci(10)
print(list(fib))
```

```python
# 6. Generator for powers of 2:
def powers_of_2(exp):
    for i in range(exp + 1):
        yield 2**i

print(list(powers_of_2(5)))  # Output: [1, 2, 4, 8, 16, 32]
```

```python
# 7. Generator for reading a file line by line:
def read_file_lines(filepath):
    with open(filepath) as f:
        for line in f:
            yield line.strip()

# Example: Uncomment below line with your file
# for line in read_file_lines("example.txt"): print(line)
```

```python
# 8. Lambda function to sort tuples by second element:
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(2, 1), (4, 2), (1, 3)]
```

```python
# 9. Convert Celsius to Fahrenheit using map():
celsius = [0, 20, 30]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)  # Output: [32.0, 68.0, 86.0]
```

```python
# 10. Remove vowels using filter()
def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

print(remove_vowels("Hello World"))  # Output: "Hll Wrld"
```

```python
# 11. Accounting routine with lambda and map:
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda x: (x[0], round(x[2] * x[3] + (10 if x[2] * x[3] < 100 else 0), 2)),
    orders
))
print(result)
```
